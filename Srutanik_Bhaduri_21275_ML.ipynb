# Importing the dataset using Pandas

```python
import pandas as pd

file_path = 'training_data.csv'

data = pd.read_csv(file_path)
pd.set_option('display.max_rows', None)

data
```

# Pre-processing and cleaning

## Converting MUTATED and NOT_MUTATED in the dataset to 1 and 0 respectively. This allows for the models to run smoothly and make it a simple classification problem.

```python
mutation_columns = [col for col in data.columns if 'MUTATED' in data[col].unique()]
for col in mutation_columns:
    data[col] = data[col].map({'MUTATED': 1, 'NOT_MUTATED': 0})
```

## Dropping rows with no data under any feature. Here no data is represented by '--'

```python
data_cleaned = data[~data.isin(['--']).any(axis=1)]
data_cleaned
```

## Converting age to a float value

```python
def convert_age_to_years_safe(age_str):
    if 'years' in age_str and 'days' in age_str:
        years, days = age_str.split(' years ')
        days = days.split(' days')[0]  
        years = age_str.split(' years')[0]
        days = 0
    else:
        years, days = 0, 0

    total_years = float(years) + float(days) / 365.25  
    return total_years

data_cleaned.loc[:, 'Age_at_diagnosis'] = data_cleaned['Age_at_diagnosis'].apply(convert_age_to_years_safe)
data_cleaned
```

## One-hot encoding of the features Primary_diagnosis, Race, Gender using pandas

```python
data_encoded = pd.get_dummies(data_cleaned, columns=['Primary_Diagnosis', 'Race','Gender'])
data_encoded
```

# The model run for the data with the following :
1. Grid search
2. Feature selection
3. Dimensionality reduction
4. Cross validation
5. Model training

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn import svm
from sklearn.linear_model import LogisticRegression, Lasso
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.metrics import f1_score, make_scorer, precision_score, recall_score, confusion_matrix
from sklearn.feature_selection import SelectFromModel
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline

class Classification:

    def __init__(self, clf_opt='rf'):
        self.clf_opt = clf_opt

    def Pipeline(self):
        feature_selection = SelectFromModel(Lasso(alpha=0.01, random_state=42))
        pca = PCA(n_components=0.95)  # Keeping 95% of the variance

        if self.clf_opt == 'rf':
            print('\nRandom Forest Classifier\n')
            clf = RandomForestClassifier(random_state=42)
            param_grid = {
                'classifier__n_estimators': [10, 20, 50, 60],
                'classifier__max_depth': [None, 10, 20, 30],
                'classifier__min_samples_split': [2, 3, 4],
                'classifier__min_samples_leaf': [1, 2, 3, 4],
            }
        elif self.clf_opt == 'dt':
            print('\nDecision Tree Classifier\n')
            clf = DecisionTreeClassifier(random_state=42)
            param_grid = {
                'classifier__criterion': ['gini', 'entropy'],
                'classifier__max_features': ['sqrt', 'log2', 0.7, None],
                'classifier__max_depth': [10, 40, 45, 60, 70, 80],
                'classifier__ccp_alpha': [0.009, 0.01, 0.05, 0.1, 0.2],
            }
        elif self.clf_opt == 'mlp':
            print('\nMLP Classifier\n')
            clf = MLPClassifier(random_state=42)
            param_grid = {
                'classifier__hidden_layer_sizes': [(10, 5, 15, 20), (15, 20, 50), (50, 70, 100)],
                'classifier__activation': ['tanh', 'relu'],
                'classifier__solver': ['sgd', 'adam'],
                'classifier__alpha': [0.001, 0.05, 0.06],
                'classifier__max_iter': [200, 300, 500, 1000],
                'classifier__learning_rate': ['constant', 'adaptive'],
            }
        elif self.clf_opt == 'svm':
            print('\nSupport Vector Machine Classifier\n')
            clf = svm.SVC(random_state=42)
            param_grid = {
                'classifier__C': [0.1, 0.5, 5, 10],
                'classifier__gamma': [0.1, 0.5, 1, 10],
                'classifier__kernel': ['rbf', 'poly', 'sigmoid']
            }
        elif self.clf_opt == 'ab':
            print('\nAdaBoost Classifier\n')
            be1 = svm.SVC(kernel='linear', class_weight='balanced', probability=True)
            be2 = LogisticRegression(solver='liblinear', class_weight='balanced')
            be3 = DecisionTreeClassifier(max_depth=50)
            
            clf = AdaBoostClassifier(algorithm='SAMME.R', n_estimators=100)
            param_grid = {
                'classifier__base_estimator': [be1, be2, be3],
                'classifier__n_estimators': [100],
                'classifier__random_state': [0, 10],
            }
        elif self.clf_opt == 'ls':
            print('\nLinear SVC Classifier\n')
            clf = svm.LinearSVC(class_weight='balanced', dual=True)
            param_grid = {
                'classifier__C': [0.1, 1, 100],
                'classifier__max_iter': [1000, 10000],
            }
        elif self.clf_opt == 'nb':
            print('\nMultinomial Naive Bayes Classifier\n')
            clf = MultinomialNB(fit_prior=True, class_prior=None)
            param_grid = {
                'classifier__alpha': [0, 1],
            }
        elif self.clf_opt == 'lr':
            print('\nLogistic Regression Classifier\n')
            clf = LogisticRegression(solver='liblinear', class_weight='balanced')
            param_grid = {
                'classifier__penalty': ['l1', 'l2'],
                'classifier__C': [0.001, 0.01, 0.1, 1, 10],
                'classifier__fit_intercept': [True, False],
                'classifier__max_iter': [100, 200, 300],
            }
        elif self.clf_opt == 'knn':
            print('\nK-Nearest Neighbors Classifier\n')
            clf = KNeighborsClassifier()
            param_grid = {
                'classifier__n_neighbors': [3, 5, 7],
                'classifier__leaf_size': [30, 50, 70],
                'classifier__p': [1, 2]
            }
        else:
            print('Invalid classifier option')
            return None, None

        pipeline = Pipeline([('feature_selection', feature_selection), ('pca', pca), ('classifier', clf)])
        return pipeline, param_grid

    def Classification(self):
        X = pd.read_csv("training_modified_1.csv")
        y = pd.read_csv("labels_1.csv")

        stratified_kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

        for train_index, test_index in stratified_kfold.split(X, y):
            X_train, X_test = X.iloc[train_index], X.iloc[test_index]
            y_train, y_test = y.iloc[train_index], y.iloc[test_index]

            pipeline, param_grid = self.Pipeline()
            if pipeline is None:
                return

            scorer = make_scorer(f1_score, average='weighted')
            grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, scoring=scorer, cv=5)

            print(f'\nTraining Classifier with {self.clf_opt}\n')
            grid_search.fit(X_train, y_train.values.ravel())

            print("Best Hyperparameters:", grid_search.best_params_)
            best_model = grid_search.best_estimator_
            y_pred = best_model.predict(X_test)

            f1_score_final = f1_score(y_test, y_pred, average='weighted')
            precision = precision_score(y_test, y_pred, average='weighted')
            recall = recall_score(y_test, y_pred, average='weighted')

            print(f"F1-Score: {f1_score_final:.2f}")
            print(f"Precision: {precision:.2f}")
            print(f"Recall: {recall:.2f}")

            cm = confusion_matrix(y_test, y_pred)
            print("Confusion Matrix:")
            print(cm)
```

## Running the best model with highest metrics

```python
from sklearn.decomposition import PCA
from sklearn.feature_selection import SelectFromModel
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import Lasso
from sklearn.pipeline import Pipeline
import pandas as pd

feature_selection = SelectFromModel(Lasso(alpha=0.01, random_state=42))
pca = PCA(n_components=0.95)  # Keeping 95% of the variance
knn_classifier = KNeighborsClassifier(n_neighbors=3, leaf_size=30, p=1)

best_classifier = Pipeline([
    ('feature_selection', feature_selection),
    ('pca', pca),
    ('knn', knn_classifier)
])

X_train = pd.read_csv("training_modified_1.csv")  # replace with your training data file path
y_train = pd.read_csv("labels_1.csv")  # replace with your training labels file path

best_classifier.fit(X_train, y_train.values.ravel())

predictions = best_classifier.predict(X_test)
```
